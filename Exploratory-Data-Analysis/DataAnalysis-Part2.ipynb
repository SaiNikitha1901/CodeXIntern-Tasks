## 2. Imputers

```python
imputer=SimpleImputer(missing_values=np.nan,strategy="mean")
```

```python
imputer.fit(df[['Salary']])
```

```python
imputer.transform(df[['Salary']])
#We can equate df[['Salary']] to apply the changes to df
```

```python
df.iloc[:,1:3].values
```

```python
#We can also use iloc and slicing and pass it to imputer since it takes only 2D elements
```

```python
df[['Age','Salary']]=imputer.fit_transform(df.iloc[:,1:3].values)
```

```python
df
```

```python
df['Age']=df['Age'].round(0).astype(int)
df['Salary']=df['Salary'].round(0).astype(float)
```

```python
df
```

```python
#Now to replace the categorical data
imputer=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
df[['Purchased']]=imputer.fit_transform(df.iloc[:,3:].values)
df[['Country']]=imputer.fit_transform(df.iloc[:,0:1].values)
```
