# Dealing with Outliers
### 1. Trimming
### 2.Capping
## Various Ways of finding an outlier:
### a)Using z score-Normal distribution b)Using Interqaurtile range-Skewed Distribution  c)Using Percentile-Other Distributions

````python
#Using Z-score
#The values outside of 3 standard deviations i.e., outside (mue-3sigma,mue+3sigma) are considered to be outliers
````

````python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
````

````python
data=pd.read_csv("Scholarship.csv")
````

````python
df=data.copy()
````

````python
df.head()
````

````python
df.shape
````

````python
# We need to check the distribution hence we plot the data 
````

### Introduction to plots:
> Think of a subplot as dividing one big graph area into smaller boxes,
so you can draw multiple charts in one window.
```python
plt.subplot(rows,columns,index)->Index refers to which index graph we are plotting but starts from 1
plt.subplot(2,1,1)->Two graphs stacked vertically (top one)
plt.plot([1,2,3],[6,4,5]) ->This is a line plot

```


````python
plt.subplot(1,2,1)
sns.histplot(df["semester_percentage"],kde=True)
plt.subplot(1,2,2)
sns.histplot(df["scholarship_exam_marks"],kde=True)
plt.show()
````

````python
#We can see that semester percentage is normally distributed and scholarship exam marks is rigt skewed
````

````python
mean=df["semester_percentage"].mean()
std=df["semester_percentage"].std()
````

````python
print(mean,std)
````

### TRIMMING

````python
Highest_allowed=mean+(3*std)
Lowest_allowed=mean-(3*std)
print(Highest_allowed,Lowest_allowed)
````

````python
df[(df["semester_percentage"]<Lowest_allowed) |(df["semester_percentage"]>Highest_allowed)]
````

````python
df[(df["semester_percentage"]>Lowest_allowed) & (df["semester_percentage"]<Highest_allowed)]
````

````python
#We havent permanently changed the dataset yet. We can just equate the df to the above value to do so.
````

> Using Z-score

````python
df["Z_score"]=(df["semester_percentage"]-df["semester_percentage"].mean())/df["semester_percentage"].std()
````

````python
df.head()
````

````python
#Now since we have standard normal distribution curve, we can say values outside (-3,3) will be considered as outliers
````

````python
df[(df["Z_score"]<-3) |(df["Z_score"]>3)]
````

````python
df[(df["Z_score"]>-3) & (df["Z_score"]<3)]
````

### CAPPING
> Basically any value above the upperlimit is replaced with upperlimit. Similarly any value lower than the lowerlimit is replaced with lowerlimt

> We use a numpy function to do this. np.where(cond,x,y)
> Here if the the condition is true x is executed else y is executed. This is similar to the ternary operator.

````python
df["semester_percentage"]=np.where(df["semester_percentage"]>Highest_allowed,Highest_allowed,np.where(df["semester_percentage"]<Lowest_allowed,Lowest_allowed,df["semester_percentage"]))
````

````python
df
````

````python
#Checking if capping worked
df[(df["semester_percentage"]<Lowest_allowed) |(df["semester_percentage"]>Highest_allowed)]
````

> Using IQR

````python
df1=data.copy()
````

````python
#We have seen that the scholarship marks have a skewed distribution. 
sns.histplot(df1["scholarship_exam_marks"],kde=True)
````

````python
# We can visualize the outliers using the boxplot
sns.boxplot(df1["scholarship_exam_marks"])
````

````python
#We can see from both the plots that it is right skewed, meaning, there are outliers present on the higher side
````

````python
Q1=df1["scholarship_exam_marks"].quantile(0.25)
Q3=df1["scholarship_exam_marks"].quantile(0.75)
````

````python
print(Q1,Q3)
````

````python
IQR=Q3-Q1
````

````python
#We also find the quantiles using the describe function
df1["scholarship_exam_marks"].describe()
````

````python
min=Q1-1.5*IQR
max=Q3+1.5*IQR
````

````python
df1[(df1["scholarship_exam_marks"]<min) | (df1["scholarship_exam_marks"]>max)]
````

### Trimming

````python
new_data=df1[(df1["scholarship_exam_marks"]>min) & (df1["scholarship_exam_marks"]<max)]
````

````python
#Comparing the before and after
plt.figure(figsize=(12, 9)) 
plt.subplot(2,2,1)
sns.histplot(df1["scholarship_exam_marks"],kde=True)
plt.subplot(2,2,2)
sns.boxplot(df1["scholarship_exam_marks"])
plt.subplot(2,2,3)
sns.histplot(new_data["scholarship_exam_marks"],kde=True)
plt.subplot(2,2,4)
sns.boxplot(new_data["scholarship_exam_marks"])
plt.tight_layout()
plt.show()
````

### Capping

````python
new_df=np.where(df1["scholarship_exam_marks"]>max,max,np.where(df1["scholarship_exam_marks"]<min,min,df1["scholarship_exam_marks"]))
new_df = pd.DataFrame({"scholarship_exam_marks": new_df})
#This step has to be done before plotting the graph since np.where returns a numpy array not a dataframe
````

````python
#Comparing the before and after
plt.figure(figsize=(12, 9)) 
plt.subplot(2,2,1)
sns.histplot(df1["scholarship_exam_marks"],kde=True)
plt.subplot(2,2,2)
sns.boxplot(df1["scholarship_exam_marks"])
plt.subplot(2,2,3)
sns.histplot(new_df["scholarship_exam_marks"],kde=True)
plt.subplot(2,2,4)
sns.boxplot(new_df["scholarship_exam_marks"])
plt.tight_layout()
plt.show()
````
