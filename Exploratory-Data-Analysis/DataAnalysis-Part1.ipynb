# Numeric data analysis

```python
#For now, we will work with df(null values removed) and do the analysis of numeric values
df['Rating']
```

## 1. Finding average Rating

```python
print("Average Rating: ",(sum(df['Rating'])/len(df['Rating'])))
```

```python
float(str(sum(df['Rating'])/9360)[:4])
```

## 2. How many apps are there with a Rating of 5?

```python
c=0
for i in df['Rating']:
    if i==5.0:
        c+=1
print(c,"applications have a rating equal to 5")
```

## 3. Range of the Rating

```python
c=0
for i in df['Rating']:
    if i>=4.0 and i<=4.5:
        c+=1
print(c,"applications have a rating between 4-4.5")
```

## 4. Average number of Reviews 

```python
#sum(df['Reviews'])-->This gave an error and revealed that Reviews is of str data type
#Hence we find a different way to compute it.
s=0
for i in df['Reviews']:
    s+=int(i)
print("Average number of reviews:",int(s/len(df['Reviews'])))
    
```

# Categorical data analysis

```python
lst1=df['Category'].unique() #To find all the unique categories
lst1
```

## 1. How many unique categories are there?

```python
for i in lst1:
    print(i)
print("Total Categories:",len(lst1))
```

## 2. Total apps in ART and DESIGN category

```python
s=0
for i in df['Category']:
    if(i=="ART_AND_DESIGN"):
        s=s+1
print(s,"apps are of ARTS and DESIGN category")
```

## 3. Total App-Types

```python
df['Type'].unique()
```

## 4. Total number of FREE and PAID apps

```python
f=0
for i in df['Type']:
    if(i=="Free"):
        f+=1
print(f,"free apps and",(len(df['Type'])-f),"paid apps")
```

## 5. Percenatge of free apps

```python
fp=(f/(len(df['Type'])))*100
print(int(fp))
```

## 6. Content Rating Categories

```python
for i in df['Content Rating'].unique():
    print(i)
```

## 7. Total Number of apps in each Category

```python
#Now if we want to print the category name and how many enteries of that category are present
for name in df['Category'].unique():
    ct=0
    for i in df['Category']:
        if(i==name):
            ct+=1;
    print(name,":",ct)   
```

```python
#We can store this in a dictnary as well
categories={}
for name in df['Category'].unique():
    ct=0
    for i in df['Category']:
        if(i==name):
            ct+=1;
    categories[name]=ct
print(categories)
```

## 8. Total number of apps in each Content Rating

```python
for name in df['Content Rating'].unique():
    ct=0
    for i in df['Content Rating']:
        if(i==name):
            ct+=1;
    print(name,":",ct)  
```
